```python
import pandas as pd
```

```python
df = pd.read_csv('Resources/nfl_data_with_win_percentages.csv')
df.head(100)
```

```python
df['cumulative_win_percent'] = df['cumulative_win_percent'].fillna('')
df.head(100)
```

```python
df_forViewing = df[[
    'Game ID',
    'schedule_season',
    'schedule_week_clean',
    'team',
    'cumulative_win_percent',
    'win_lose_draw'
]]

df_forViewing.head(100)
```

```python
# Initalize game_count and win_count variables
game_count = 0
win_count = 0

# Iterate through each row in the dataframe
for index, row in df.iterrows():
    
    # First game of the season, we chose to use win percent of 0.5
    if index == 0:
        df.loc[index, 'cumulative_win_percent'] = 0.5
    
    else:
        
        # If team in subsequent row is same as previous, increment game_count
        # and increment win_count (if result of matchup is a win)
        if df.loc[index, 'team'] == df.loc[index - 1, 'team']:
            game_count += 1
            if df.loc[index - 1, 'win_lose_draw'] == 1:
                win_count += 1
            
            # win percent = win_count / game_count
            df.loc[index, 'cumulative_win_percent'] = win_count / game_count
        
        # If team in subsequent row is a new team, re-initialize
        else:
            game_count = 0
            win_count = 0
            df.loc[index, 'cumulative_win_percent'] = 0.5

df.head(100)
```

```python
df.to_csv('Resources/nfl_data_with_final_win_percentages.csv')
```
